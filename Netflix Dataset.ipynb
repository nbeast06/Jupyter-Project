```python
df = df.loc[df['type'] == 'TV Show']
df.head()
```

```python
def splitstack(df,col):
    #split the row and create stack
    s = df[col].str.split(',', expand=True).stack()
    #get the indices of 0 level of stack, we have 0 and 1
    i = s.index.get_level_values(0)
    #expand the dataframe based on the indices of stack
    df2 = df.loc[i].copy()
    #replace stacked column
    df2[col] = s.values
    #strip/trim spaces at edges of data
    df2[col] =df2[col].str.strip()\
    #reset index
    df2 = df2.reset_index()
    return df2

df2 = splitstack(df,'listed_in')
df2 = splitstack(df2,'cast')

df2.head(100)
```
